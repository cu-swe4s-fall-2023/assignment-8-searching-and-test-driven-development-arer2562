```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Agrofood_co2_emission.csv',sep=',')
print(df.head())
```

```python
zdf = df[df['Area'] == 'Brazil']
cdf = df[df['Area'] == 'Canada']

fig, ax = plt.subplots()
plt.plot(zdf['Year'],zdf['Fires in humid tropical forests'], label='Brazil')
plt.plot(cdf['Year'],cdf['Fires in humid tropical forests'], label='Canada')
ax.legend(loc='upper right')
ax.set_xlabel('Year')
ax.set_ylabel('Fires in humid tropical forests')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(zdf['Savanna fires'], bins=20, edgecolor='black')
ax.set_xlabel('Savanna fires')
ax.set_ylabel('Freq.')
ax.set_title('Freq. Savanna Fires in Brazil', loc='center')
ax.spines[['right', 'top']].set_visible(False)
print(ax.get_xlim())
ax.hlines(y=3,xmin=ax.get_xlim()[0],xmax=ax.get_xlim()[1], color='red',linestyle='--')
plt.show()
```

```python
import numpy as np
fig, ax = plt.subplots()
ax.scatter(zdf['Pesticides Manufacturing'],zdf['total_emission'],alpha=.5,s=50,c='purple')
plt.plot(np.unique(zdf['Pesticides Manufacturing']), np.polyd(np.polyfit(zdf['Pesticides Manufacturing'], zdf[;total_emission'],1)
plt.show()
```

## Multipanel plots


```python
import seaborn as sns
import numpy as np
fig, axes = plt.subplots(2,2, width_ratios=[3,1],height_ratios=[2,1])
fig.set_size_inches(10,10)
print(axes.shape)
axes[0,0].scatter(zdf['Pesticides Manufacturing'],zdf['total_emission'],alpha=.5,s=50,c='purple')
axes[0,0].set_title('A',loc='left')
axes[0,0].set_xlabel('Pesticides Manufacturing')
axes[0,0].set_ylabel('Total emission')

axes[0,1].boxplot(zdf['Pesticides Manufacturing'])
axes[0,1].set_title('B\n',loc='left')
axes[0,1].spines[['right','top']].set_visible=(False)
axes[0,1].set_ylabel('Pesticides Manufacturing')

axes[1,0].violinplot(zdf['total_emission'])

sns.kdeplot(np.array(zdf['total_emission']), bw=0.5,ax=axes[1,1])
plt.savefig('histogram.png',dpi=300)
plt.show()
plt.show()
```
